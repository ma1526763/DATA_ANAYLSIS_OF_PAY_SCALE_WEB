```python
import pandas
import re

df = pandas.read_csv("Py Scale.csv")

# cleaning data 
df['Early Career Pay'] = df['Early Career Pay'].replace('[\$,]', '', regex=True)
df['Early Career Pay'] = pandas.to_numeric(df['Early Career Pay'])
df['Mid Career Pay'] = df['Mid Career Pay'].replace('[\$,]', '', regex=True)
df['Mid Career Pay'] = pandas.to_numeric(df['Mid Career Pay'])
df['% High Meaning'] = df['% High Meaning'].replace('[\%]', '', regex=True)



# highest degree with early Career Pay
early_highest_pay_id = df['Early Career Pay'].idxmax()
print(df.loc[early_highest_pay_id])

# top 5 highest major degrees with Early Career Pay 
early_highest_pay = df.sort_values('Early Career Pay', ascending=False)
print(early_highest_pay.head())

# lowest 5 major degrees with Early Career Pay 
early_highest_pay = df.sort_values('Early Career Pay')
print(early_highest_pay.head())


# sorting based on Early Career Pay 
# ascending
sorting_based_on_early_career = df.sort_values('Early Career Pay')
print(sorting_based_on_early_career)

# descending
sorting_based_on_early_career = df.sort_values('Early Career Pay', ascending=False)
print(sorting_based_on_early_career)


# highest degree with Mid Career Pay
mid_highest_pay_id = df['Mid Career Pay'].idxmax()
print(df.loc[mid_highest_pay_id])

# top 5 highest major degrees with Mid Career Pay
mid_highest_pay = df.sort_values('Mid Career Pay', ascending=False)
print(mid_highest_pay.head())

# lowest 5 major degrees with Mid Career Pay
mid_highest_pay = df.sort_values('Mid Career Pay')
print(mid_highest_pay.head())


# sorting based on Mid Career Pay
# ascending
sorting_based_on_mid_career = df.sort_values('Mid Career Pay')
print(sorting_based_on_mid_career)

# descending
sorting_based_on_mid_career = df.sort_values('Mid Career Pay', ascending=False)
print(sorting_based_on_mid_career)



# top 5 highest major degrees with Highest Meaning
highest_meaning = df.sort_values('% High Meaning', ascending=False)
print(highest_meaning.head())

# lowest 5 major degrees with Highest Meaning
highest_meaning = df.sort_values('% High Meaning')
print(highest_meaning.head())


# sorting based on Highest Meaning
# ascending
sorting_based_on_highest_meaning = df.sort_values('% High Meaning')
print(sorting_based_on_highest_meaning)

# descending
sorting_based_on_highest_meaning = df.sort_values('% High Meaning', ascending=False)
print(sorting_based_on_highest_meaning)

# group based on Degree Type
print(df.groupby('Degree Type').count())



```
